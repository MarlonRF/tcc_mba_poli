```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar a tabela
# Substitua 'dados.xlsx' pelo caminho do arquivo ou ajuste para seu formato de dados
df = pd.read_excel('dados.xlsx')

# Função para analisar valores únicos, intervalos e faltantes
def analyze_columns(dataframe, n_unicos=10, valores_validos=None):
    analysis = []
    for col in dataframe.columns:
        missing_count = dataframe[col].isnull().sum()
        missing_percentage = (missing_count / len(dataframe)) * 100
        if dataframe[col].dtype == 'object' or dataframe[col].nunique() < 50:
            # Variáveis categóricas ou com poucos valores únicos
            unique_values = dataframe[col].dropna().unique()

            analysis.append({
                'Coluna': col,
                'Tipo': 'Categórica',
                'Valores ausentes': missing_count,
                '/% ausentes': round(missing_percentage, 2),
                'Valores únicos': unique_values[:n_unicos],  # Mostra os primeiros 10 valores únicos
                'Contagem únicos': len(unique_values)
            })

            if any valores_validos not in valores_validos
                


        
        else:
            # Variáveis numéricas ou contínuas

            analysis.append({
                'Coluna': col,
                'Tipo': 'Numérica',
                'Missing Count': missing_count,
                '/% ausentes': round(missing_percentage, 2),
                'Min': dataframe[col].min(),
                'Max': dataframe[col].max(),
                'Mean': dataframe[col].mean(),
                'Contagem únicos': dataframe[col].nunique()
            })
    return pd.DataFrame(analysis)

# Analisar o DataFrame
analysis_df = analyze_columns(df)

# Função para verificar valores ausentes
def valor_ausente(valor):
    return valor > 0

# Função para validar classes permitidas
def classes_permitidas(valor, classes_permitidas):
    return valor not in classes_permitidas

# Função para validar intervalo permitido
def intervalo_permitido(valor, intervalo_permitido):
    min_val, max_val = intervalo_permitido
    return not (min_val <= valor <= max_val)

# Função para verificar tipos esperados de colunas com base em um dicionário
def tipos_esperados(dataframe, dict_tipos):
    results = {}
    for col, expected_type in dict_tipos.items():
        if col in dataframe.columns:
            results[col] = all(isinstance(val, expected_type) for val in dataframe[col].dropna())
        else:
            results[col] = False  # Coluna não encontrada
    return results


def dicionarioColunasNormalizadas(df):
    cabecalhos = dict(zip(list(df.columns),[str(nome).lower() for nome in list(df.columns)]))
    return cabecalhos


import pandas as pd

def apply_column_types(df, type_dict):
    """
    Apply data types to a DataFrame based on a dictionary, using pd.to_numeric
    for numeric types with coercion when necessary.

    Args:
        df (pd.DataFrame): The DataFrame to process.
        type_dict (dict): A dictionary mapping column names to desired data types.

    Returns:
        pd.DataFrame: The DataFrame with updated column types.
    """
    for col, dtype in type_dict.items():
        if col in df.columns:
            try:
                if dtype == 'datetime64[ns]':
                    # Convert to datetime, coercing invalid entries to NaT
                    df[col] = pd.to_datetime(df[col], errors='coerce')
                elif dtype == 'float64' or dtype == 'Int64':
                    # Convert to numeric, coercing invalid entries to NaN
                    df[col] = pd.to_numeric(df[col], errors='coerce')
                    # For Int64, convert from float to nullable integer type
                    if dtype == 'Int64':
                        df[col] = df[col].astype('Int64')
                else:
                    # Apply general type conversion
                    df[col] = df[col].astype(dtype)
            except Exception as e:
                print(f"Error converting column {col} to {dtype}: {e}")
        else:
            print(f"Column {col} not found in DataFrame.")
    return df


    


def verificar_coluna_conforme_dicionario(df, coluna, dict_valores):
    """
    Verifica se os valores de uma coluna do DataFrame estão de acordo com as regras no dicionário.
    Para colunas categóricas, os valores devem estar na lista permitida.
    Para colunas numéricas, os valores devem estar dentro de um intervalo definido.

    Args:
        df (pd.DataFrame): O DataFrame a ser verificado.
        coluna (str): Nome da coluna a ser verificada.
        dict_valores (dict): Dicionário contendo as regras de validação:
            - Categórica: {'coluna': [valores_permitidos]}
            - Numérica: {'coluna': [min_val, max_val]}

    Returns:
        tuple: (bool, list)
            - True se todos os valores estão de acordo com as regras.
            - Lista com os valores fora das regras.
    """
    # Obter os valores únicos da coluna
    valores_unicos = df[coluna].unique()

    # Verificar se a coluna é categórica ou numérica
    regras = dict_valores.get(coluna)
    if len(regras) == 2 and isinstance(regras[0], (int, float)) and isinstance(regras[1], (int, float)):
        # Coluna numérica
        min_val, max_val = regras
        valores_fora = [valor for valor in valores_unicos if valor < min_val or valor > max_val]
    else:
        # Coluna categórica
        valores_fora = [valor for valor in valores_unicos if valor not in regras]

    # Retornar o resultado
    return (len(valores_fora) == 0, valores_fora)
```
